## Ordenação

### Model 1

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
quintis = rf_model1.index
events_num = rf_model1['Qtd de eventos (Target)']
ax.bar(quintis,events_num)

plt.xlabel('Quintis')
plt.ylabel('Qtd de Eventos por Quintil')
plt.title("Ordenação do Modelo 1: Random Forest + FS1")

# Make some labels.
labels = events_num.values
rects = ax.patches
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2, height + 2.5, label, ha="center", va="bottom"
    )
plt.savefig('../reports/figures/ordenation_model1_rf_fs1.png', bbox_inches='tight', dpi=plt.gcf().dpi)
plt.show()
```
