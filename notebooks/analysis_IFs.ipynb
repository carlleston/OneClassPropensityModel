```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
if_model5= pd.read_csv("results_model5_iforest_fs1.csv",sep=';', index_col = 0)
if_model6= pd.read_csv("results_model6_iforest_fs2.csv",sep=';', index_col = 0)
if_model7= pd.read_csv("results_model7_iforest_fs3.csv",sep=';', index_col = 0)
if_model8= pd.read_csv("results_model8_iforest_fs4.csv",sep=';', index_col = 0)
```

```python
# plt.plot(np.cumsum(df_rfbase['% de eventos no Decil']), label='% de eventos')
plt.plot(if_model5['KS'], label='KS IF+FS1')
plt.plot(if_model6['KS'], label='KS IF+FS2')
plt.plot(if_model7['KS'], label='KS IF+FS3')
plt.plot(if_model8['KS'], label='KS IF+FS4')


plt.xlabel('Quintis')
plt.ylabel('KS (%)')
plt.title("Comparação dos modelos: KS Isolation Forest")
plt.xticks(range(1,6))
plt.legend()
plt.savefig('../reports/figures/compare_model_ks_if.png', bbox_inches='tight', dpi=plt.gcf().dpi)
```

## Ordenação

### Model 5

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
quintis = if_model5.index
events_num = if_model5['Qtd de eventos (Target)']
ax.bar(quintis,events_num)

plt.xlabel('Quintis')
plt.ylabel('Qtd de Eventos por Quintil')
plt.title("Ordenação do Modelo 5: Isolation Forest + FS1")

# Make some labels.
labels = events_num.values
rects = ax.patches
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2, height, label, ha="center", va="bottom"
    )
plt.savefig('../reports/figures/ordenation_model5_if_fs1.png', bbox_inches='tight', dpi=plt.gcf().dpi)
plt.show()
```
