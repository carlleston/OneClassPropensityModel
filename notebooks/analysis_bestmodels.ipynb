```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rf_model3= pd.read_csv("results_model3_rf_fs4.csv",sep=';', index_col = 0)
if_model8= pd.read_csv("results_model8_iforest_fs4.csv",sep=';', index_col = 0)
ocsvm_model9= pd.read_csv("results_model9_ocsvm.csv",sep=';', index_col = 0)
```

```python
# plt.plot(np.cumsum(df_rfbase['% de eventos no Decil']), label='% de eventos')
plt.plot(rf_model3['KS'], label='KS RF+FS4')
plt.plot(if_model8['KS'], label='KS IF+FS4')
plt.plot(ocsvm_model9['KS'], label='KS OCSVM+FS1')


plt.xlabel('Quintis')
plt.ylabel('KS (%)')
plt.title("Comparação dos modelos: KS")
plt.xticks(range(1,6))
plt.legend()
plt.savefig('../reports/figures/compare_model_ks_geral.png', bbox_inches='tight', dpi=plt.gcf().dpi)
```

```python
# plt.plot(np.cumsum(df_rfbase['% de eventos no Decil']), label='% de eventos')
plt.plot(rf_model3['KS'][0:3], label='KS RF+FS4')
plt.plot(if_model8['KS'][0:3], label='KS IF+FS4')
plt.plot(ocsvm_model9['KS'][0:3], label='KS OCSVM+FS1')


plt.xlabel('Quintis')
plt.ylabel('KS (%)')
plt.title("Comparação dos modelos: KS")
plt.xticks(range(1,4))
plt.legend()
plt.savefig('../reports/figures/compare_model_ks_geral_zoom3quin.png', bbox_inches='tight', dpi=plt.gcf().dpi)
```
