```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ocsvm_model9= pd.read_csv("results_model9_ocsvm.csv",sep=';', index_col = 0)
ocsvm_model10= pd.read_csv("results_model10_ocsvm_fs4.csv",sep=';', index_col = 0)
```

## Ordenação

### Model 9

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
quintis = ocsvm_model9.index
events_num = ocsvm_model9['Qtd de eventos (Target)']
ax.bar(quintis,events_num)

plt.xlabel('Quintis')
plt.ylabel('Qtd de Eventos por Quintil')
plt.title("Ordenação do Modelo 9: OCSVM + FS1")

# Make some labels.
labels = events_num.values
rects = ax.patches
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2, height, label, ha="center", va="bottom"
    )
plt.savefig('../reports/figures/ordenation_model9_OCSVM_fs1.png', bbox_inches='tight', dpi=plt.gcf().dpi)
plt.show()
```
